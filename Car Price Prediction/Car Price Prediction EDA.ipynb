```python
power= []
for i in df.Max_Power:
    if str(i).endswith('bhp'):
        i = i[:-3]
        power.append(float(i))
    elif str(i).endswith('Nm'):
        i = np.nan
        power.append(i)
    elif str(i).endswith('nm'):
        i = np.nan
        power.append(i)    
    elif str(i).endswith('PS'):
        i=i[:-2]
        i=round((float(i)/1.014),2)
        power.append(float(i))
    elif str(i).endswith('ps'):
        i=i[:-2]
        i=round((float(i)/1.014),2)
        power.append(float(i))    
    elif str(i).endswith('rpm'):
        i=i[:-10]
        power.append(i) 
    elif str(i).endswith('RPM'):
        i=i[:-14]
        i=round((float(i)/1.014),2)
        power.append(float(i))     
    elif str(i).endswith(''):
        i=i
        power.append(i)        
```

```python
power
```

```python
df['Max_Power']=power
```

```python
df['Max_Power'].unique()
```

```python
power=[]
for i in df.Max_Power:
    if str(i).endswith('PS a'):
        i = i[:-4]
        i=round((float(i)/1.014),2)
        power.append(float(i))
    elif str(i).endswith('PS '):
        i = i[:-3]
        i=round((float(i)/1.014),2)
        power.append(float(i))
    elif str(i).endswith('PS at'):
        i = i[:-5]
        i=round((float(i)/1.014),2)
        power.append(float(i))  
    elif str(i).endswith('bhp a'):
        i = i[:-5]
        power.append(i)
    elif str(i).endswith('Bhp'):
        i = i[:-3]
        power.append(i)    
    elif str(i).endswith(''):
        i = i
        power.append(i)
```

```python
torque=[]
for i in df.Torque:
    if str(i).endswith('Nm'):
        i = i[:-2]
        torque.append(i)
    elif str(i).endswith('nm'):
        i = i[:-2]
        torque.append(i)
    elif str(i).endswith('NM'):
        i = i[:-2]
        torque.append(i)  
    elif str(i).endswith('rpm'):
        i = i[0:6]
        torque.append(i)
    elif str(i).endswith('RPM'):
        i = i[0:6]
        torque.append(i)    
    elif str(i).endswith(''):
        i = i
        torque.append(i)    
```

```python
torque
```

```python
df['Torque']=torque
```

```python
df['Torque'].unique()
```

```python
torque=[]
for i in df.Torque:
    if str(i).endswith('k'):
        i = i[:-1]
        i=round((float(i)*9.80665),2)
        torque.append(float(i))
    elif str(i).endswith('K'):
        i = i[:-1]
        i=round((float(i)*9.80665),2)
        torque.append(float(i))
    elif str(i).endswith('Nm '):
        i = i[:-3]
        torque.append(i)
    elif str(i).endswith('Nm'):
        i = i[:-2]
        torque.append(i)    
    elif str(i).endswith('kgm'):
        i = i[:-3]
        i=round((float(i)*9.80665),2)
        torque.append(float(i))
    elif str(i).endswith('kg-m'):
        i = i[:-4]
        i=round((float(i)*9.80665),2)
        torque.append(float(i))    
    elif str(i).endswith(''):
        i = i
        torque.append(i)    
```
